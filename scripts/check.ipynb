```python
import pandas as pd
```

```python
df1 = pd.read_csv('lsgi-election-kerala-2010.csv')
df2 = pd.read_csv('lsgi-election-kerala-2015.csv')
df3 = pd.read_csv('lsgi-election-kerala-2015-fix.csv')
df4 = pd.read_csv('lsgi-election-kerala-2020.csv')

df = pd.concat([df1, df2, df3, df4])
df
```

```python
df.to_csv('lsgi-election-kerala.csv', index=False)
```
